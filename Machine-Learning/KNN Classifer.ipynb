# KNN Classifier


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df=pd.read_csv("mobile_clean.csv")
```

```python
df.head()
```

```python
X=df.drop('is_liked',axis=1).values
Y=df['is_liked']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify = Y,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler #scale doen karta hai from 0-1
sc=StandardScaler()
Xtrain=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
accuracy=0
for i in range(1,40): # to decide number of neighbours
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    Y_pred=knn.predict(X_test)
    acc=accuracy_score(Y_pred,Y_test) # accuracy score as to how good the model is
    #if the accuracy is better than the last then a new best accuracy is set
    if acc>accuracy:
        accuracy=acc
        bst_i=i #
print(accuracy)
print(bst_i)
```
